```python
fuel_efficiency = ['excellent', 'good', 'med', 'crap']
maintenance = ['vhigh', 'high', 'med', 'low']
doors = ['two', 'three', 'four', 'many']
type = ['sports', 'sedan', 'wagon']
size = ['small', 'med', 'big']
safety = ['low', 'med', 'high']
```

```python
# function definition: getting the counts (frequencies of occurrence)
def count(data,colname,label,target):    # for whichever data set, pick a column and one value in that column
                                         # and an output label (target) to compare against
    condition = (data[colname] == label) & (data['RATING'] == target)
    print(data[condition])
    return len(data[condition])
```

```python
# list of predicted values
predicted = []

# dictionary to store probabilities
probabilities = {0:{},1:{}}

#length of current training set
train_len = int(splitRatio * len(shuffCarData))
print(train_len)

#splitting the data into training and test sets
training = shuffCarData.iloc[:train_len,:] # the first train_len rows, all columns
```

```python
test_X = shuffCarData.iloc[train_len:,:-1] # features in the test set
test_y = shuffCarData.iloc[train_len:,-1]  # targets (the last entry in each row; i.e., the last column)
```

```python
# Get the unique target labels in the 'RATING' column
target_labels = training['RATING'].unique()

# Initialize dictionaries to store counts and priors
counts = {}
priors = {}

# Compute counts and priors for each target label
for label in target_labels:
    counts[label] = count(training, 'RATING', label, label)  # Count occurrences of each label
    priors[label] = counts[label] / len(training)  # Calculate prior probability

# Print the results
for label in target_labels:
    print(f"The prior '{label}' count is", counts[label])
    print(f"The prior probability of '{label}' is", priors[label])

```

```python
# Calculate the counts for each class in the training set
count_JUNK = count(training, 'RATING', 'JUNK', 'JUNK')
count_OKAY = count(training, 'RATING', 'OKAY', 'OKAY')
count_DANK = count(training, 'RATING', 'DANK', 'DANK')
count_good = count(training, 'RATING', 'good', 'good')

# Initialize the probabilities for each class (JUNK, OKAY, DANK, good) for each feature
probabilities = [{}, {}, {}, {}]  # For JUNK, OKAY, DANK, good

# TRAINING: learning the priors and class-conditionals from the training set
for col in training.columns[:-1]:  # Exclude the 'RATING' column
    probabilities[0][col] = {}  # For JUNK
    probabilities[1][col] = {}  # For OKAY
    probabilities[2][col] = {}  # For DANK
    probabilities[3][col] = {}  # For good

    if col == "fuel_efficiency":
        for k in fuel_efficiency:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(fuel_efficiency))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(fuel_efficiency))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(fuel_efficiency))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(fuel_efficiency))

    elif col == "maintenance":
        for k in maintenance:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(maintenance))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(maintenance))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(maintenance))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(maintenance))

    # Repeat similar blocks for other columns like 'doors', 'type', 'size', and 'safety'
    elif col == "doors":
        for k in doors:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(doors))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(doors))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(doors))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(doors))

    elif col == "type":
        for k in type:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(type))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(type))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(type))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(type))

    elif col == "size":
        for k in size:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(size))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(size))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(size))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(size))

    elif col == "safety":
        for k in safety:
            count_k_JUNK = count(training, col, k, 'JUNK')
            count_k_OKAY = count(training, col, k, 'OKAY')
            count_k_DANK = count(training, col, k, 'DANK')
            count_k_good = count(training, col, k, 'good')

            probabilities[0][col][k] = (count_k_JUNK + 0.001) / (count_JUNK + 0.001 * len(safety))
            probabilities[1][col][k] = (count_k_OKAY + 0.001) / (count_OKAY + 0.001 * len(safety))
            probabilities[2][col][k] = (count_k_DANK + 0.001) / (count_DANK + 0.001 * len(safety))
            probabilities[3][col][k] = (count_k_good + 0.001) / (count_good + 0.001 * len(safety))
```

```python
# Calculate the counts for each class in the training set
count_JUNK = count(training, 'RATING', 'JUNK', 'JUNK')
count_OKAY = count(training, 'RATING', 'OKAY', 'OKAY')
count_DANK = count(training, 'RATING', 'DANK', 'DANK')
count_good = count(training, 'RATING', 'good', 'good')

# Calculate the priors by dividing each count by the total number of samples
total_samples = len(training)

prior_JUNK = count_JUNK / total_samples
prior_OKAY = count_OKAY / total_samples
prior_DANK = count_DANK / total_samples
prior_good = count_good / total_samples

# Now you can use these priors in your testing loop:
for row in range(len(test_X)):
    # Initialize probabilities for each class using the priors
    prod_JUNK = prior_JUNK
    prod_OKAY = prior_OKAY
    prod_DANK = prior_DANK
    prod_good = prior_good

    # Iterate through each feature in the test set
    for feature in test_X.columns:
        feature_value = test_X[feature].iloc[row]

        # Check if the feature value exists in the probabilities dictionary for each class
        if feature_value in probabilities[0][feature]:
            prod_JUNK *= probabilities[0][feature][feature_value]
        else:
            prod_JUNK *= 0  # or some default value if the feature value is not found

        if feature_value in probabilities[1][feature]:
            prod_OKAY *= probabilities[1][feature][feature_value]
        else:
            prod_OKAY *= 0  # or some default value if the feature value is not found

        if feature_value in probabilities[2][feature]:
            prod_DANK *= probabilities[2][feature][feature_value]
        else:
            prod_DANK *= 0  # or some default value if the feature value is not found

        if feature_value in probabilities[3][feature]:
            prod_good *= probabilities[3][feature][feature_value]
        else:
            prod_good *= 0  # or some default value if the feature value is not found

    # Predict the class with the highest probability
    if prod_JUNK > max(prod_OKAY, prod_DANK, prod_good):
        predicted.append('JUNK')
    elif prod_OKAY > max(prod_JUNK, prod_DANK, prod_good):
        predicted.append('OKAY')
    elif prod_DANK > max(prod_JUNK, prod_OKAY, prod_good):
        predicted.append('DANK')
    else:
        predicted.append('good')

# Initialize correct and incorrect counts outside the loop
correct = 0
incorrect = 0

for j in range(len(test_y)):
    if predicted[j] == 'JUNK':
        if test_y.iloc[j] == 'JUNK':
            correct += 1
        else:
            incorrect += 1
    elif predicted[j] == 'OKAY':
        if test_y.iloc[j] == 'OKAY':
            correct += 1
        else:
            incorrect += 1
    elif predicted[j] == 'DANK':
        if test_y.iloc[j] == 'DANK':
            correct += 1
        else:
            incorrect += 1
    else:  # predicted is 'good'
        if test_y.iloc[j] == 'good':
            correct += 1
        else:
            incorrect += 1

# Calculate and print accuracy
accuracy = correct / (correct + incorrect)
print(f'The accuracy is {accuracy:.4f}')
```
