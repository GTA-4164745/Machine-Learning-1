```python
import numpy as np

# Given data
x = np.array([1, 3, 5])   # x values
t = np.array([4.8, 11.3, 17.2])  # t values

# Compute necessary components for the equations
xt = x * t  # element-wise multiplication x_i * t_i
x_squared = x**2  # element-wise square (x_i)^2

# Averages
x_mean = np.mean(x)
t_mean = np.mean(t)
xt_mean = np.mean(xt)
x_squared_mean = np.mean(x_squared)

# Calculate w1* and w0*
w1_star = (xt_mean - x_mean * t_mean) / (x_squared_mean - x_mean**2)
w0_star = t_mean - w1_star * x_mean

# Output the results
w0_star, w1_star
```
